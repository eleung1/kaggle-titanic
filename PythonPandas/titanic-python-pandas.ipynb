# Cleaning the data

## Transform dataframe into the shape for machine learning

```python
# Adding a column
df['Gender'] = 4
```

```python
df.head(3)
```

```python
# Gender = 4 doesn't mean anything... let's make it the first initial of the gender. M=male, F=female
#x[0] is the first char of a string
df['Gender'] = df['Sex'].map( lambda x: x[0].upper() )
```

```python
df.head(3)
```

```python
# but for machine learning, it's better to have numeric values.
df['Gender'] = df['Sex'].map( {'female':0, 'male':1} ).astype(int)
```

```python
df.head(3)
```

```python
# Do something for the embarked values

# First find out the possible values
df['Embarked'].unique()
```

```python
df['EmbarkedVal'] = df['Embarked'].map( {'S':1, 'C':2, 'Q':3} )
df.head(3)
```

# Missing values

## Most machine learning will need a complete set of values in the column to use it.

```python
median_ages = np.zeros((2,3))
median_ages
```

## Find median age of each class

```python
for i in range(0,2):
    for j in range(0,3):
        median_ages[i,j] = df[(df['Gender'] == i) & (df['Pclass'] == j+1)]['Age'].dropna().median()
```

```python
median_ages
```

## Good practice to create a new column instead of overwritting the existing one

```python
df['AgeFill'] = df['Age']
df.head()
```

## Take a look at the rows with missing age

```python
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
```

```python
# View the missing values again
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

## Create a column to indicate if the age was originally missing

```python
df['AgeIsNull'] = pd.isnull(df.Age).astype(int)
```

```python
df.head()
```

```python
# Take a look at the summary stats again
df.describe()
```

# Feature engineering

## Parch is number of parents or children onboard, and SibSp is the number of siblings or spouses onboard.  Together they represent the family size

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
df.head(3)
```

## Artificial features that we think might be advantageous to our machine learning algorithm.

```python
# Pclass has a large effect on survival, so does age.  3rd class and old people tend to not survive...
df['Age*Class'] = df.AgeFill * df.Pclass
```
